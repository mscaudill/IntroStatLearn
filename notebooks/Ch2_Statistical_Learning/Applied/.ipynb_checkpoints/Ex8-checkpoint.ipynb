Make a new qualitiative variable, called Elite, by binning the Top10perc variable. Divide universities into two groups based on whether or not the proportion of students coming from the top 10% of their classes exceeds 50%

```python
college['Elite'] = False # make a new column called Elite intialized with False
```

```python
college.head(3)['Elite']
```

## (c) vi. Further exploration

Lets examine the elite schools more closely.

```python
college['AcceptRate'] = college.Accept/college.Apps
```

```python
college.describe() # examine acceptance rates
```

Examine the schools with the minimum and maximum acceptance rates - Princeton and Emporia State universities

```python
college.loc[college['AcceptRate'].argmin()]
```

```python
college.loc[college['AcceptRate'].argmax()]
```

```python
college.boxplot(column='AcceptRate',by='Elite', figsize=(7,7));
plt.ylim(0, 1.0);
```

Examinig the boxplot we can see that the median acceptance rate for Elite colleges (53%) is 26 percentage points lower than the median acceptance rate for non-Elite colleges (79%).

```python
college[college['Elite'] == True]['AcceptRate'].median()
```

```python
college[college['Elite'] == False]['AcceptRate'].median()
```

Lets look at the Expenditures per year at elite vs. non-elite

```python
college.boxplot(column='Expend',by='Elite', figsize=(7,7));
```

So an average of about 5,000 more per year at an Elite.

We can also look at the percentage that enroll for elite and non-elites

```python
college['EnrollRate'] = college.Enroll/college.Accept
```

```python
college.boxplot(column='EnrollRate',by='Elite', figsize=(7,7));
```

This is interesting, even though the elite schools have a lower acceptance rate (and thus more exclusive) this does not translate into a higher enrollment rate. Some thoughts on this could be that added cost leads to students not taking up the offer, or that the best students likely receive multiple offers so decrease the enrollement rate at elite universities.
