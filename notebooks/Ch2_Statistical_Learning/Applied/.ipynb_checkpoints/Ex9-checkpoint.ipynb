## (a) Quantitative vs Qualitative predictors

The quantitative predictors are: mpg, cylinders, displacement, horsepower, weight, acceleration, year. The qualitative predictor is origin. The name is the index.

## (b) Ranges of Quantitative predictors 

```python
set(Auto.columns)
```

```python
quantitative = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'year']
```

```python
qualitative = list(set(Auto.columns) - set(quantitative))
```

```python
qualitative
```

```python
Auto[quantitative].apply(lambda x: x.max()-x.min() )
```

## (c) Mean and Std of Quantitative Predictors 

```python
Auto[quantitative].mean()
```

```python
Auto[quantitative].std()
```

```python
reducedAuto = pd.concat([Auto[0:9], Auto[85:]])
```

```python
reducedAuto[quantitative].apply(lambda x: x.max() - x.min())
```

```python
reducedAuto[quantitative].std()
```

## (d) Investigate predictors graphically

```python
from pandas.tools.plotting import scatter_matrix
```

```python
discrete = ['cylinders', 'year']
```

```python
continuous = list(set(quantitative) - set(discrete))
```

```python
scatter_matrix(Auto[continuous], alpha=0.5, figsize=(10,12));
```

**If we were to construct a model to predict mpgs some of the important predictors to consider would be  weight, horsepower, and displacement. Each plot of the mpg vs. predictor for these three predictors is a nonlinear decreasing trend. To a lesser degree we could also look at the number of cylinders and the acceleration as each of these has a weak trend with mpgs.**
